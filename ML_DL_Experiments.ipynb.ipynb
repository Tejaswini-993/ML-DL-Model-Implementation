```python
import numpy as np
from scipy.linalg import lu,inv
#solve a system of linear equation(ax=b)
a=np.array([[3,1,-1],[2,4,1],[-1,2,5]])
b=np.array([4,1,1])
x=np.linalg.solve(a,b)
print("Solution to ax=b:")
print(x)

#eigenvalues and eigenvectors
eigenvalues,eigenvectors=np.linalg.eig(a)
print("\nEigenvalues of a:")
print(eigenvalues)
print("\nEigenvectors of a:")
print(eigenvectors)

#inverse matrix
a=np.array([[3,1],[2,4]])
a_inv=np.linalg.inv(a)
print("\nInverse of a:")
print(a_inv)

#LU Decomposition
a=np.array([[3,1],[2,4]])
P,L,U=lu(a)
print("\nLU decomposition of a")
print("L(lower triangular matrix)")
print(L)
print("U(upper triangular matrix)")
print(U)

#Determinant of a matrix
det_A=np.linalg.det(a)
print("\nDeterminant of a:")
print(det_A)
```

```python
import numpy as np
#Define a square matrix
A=np.array([[3,1,4],[0,2,6],[0,0,5]])
eigenvalues,eigenvectors=np.linalg.eig(A)
print("Eigenvalues:\n",eigenvalues)
print("\nEigenvectors:\n",eigenvectors)



Lambda=np.diag(eigenvalues)
Q=eigenvectors
Q_inv=np.linalg.inv(Q)
print("\nQ:\n",Q)
print("\nQ_inv:\n",Q_inv)
print("\nLambda:\n",Lambda)

reconstructed_A=Q@Lambda@Q_inv
print("\nReconstructed Matrix:\n",reconstructed_A)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the pixel values to the range [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshaping the data to add a channel dimension (for grayscale)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)

# Display a few training images
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i].reshape(28, 28, 1), cmap='OrRd')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.show()

model = models.Sequential([
    # Convolutional Layer 1
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),  # MaxPooling Layer
    # Convolutional Layer 2
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    # Flatten the feature maps into a vector
    layers.Flatten(),
    # Fully Connected Layer
    layers.Dense(128, activation='relu'),
    # Output Layer
    layers.Dense(10, activation='softmax')
])

# Display the Architecture
model.summary()

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=50,
    height_shift_range=25,
    zoom_range=10,
    horizontal_flip=True,
    vertical_flip=False #vertical flip makes no sense for digits
)

datagen.fit(x_train)

# Train the model with augmented data
history = model.fit(datagen.flow(x_train, to_categorical(y_train), batch_size=32),
                    epochs=10,
                    validation_data=(x_test, to_categorical(y_test)))

test_loss, test_acc = model.evaluate(x_test, to_categorical(y_test))
print("Test accuracy:", test_acc)

# predict on test data
predictions = model.predict(x_test[:5])
predicted_labels = tf.argmax(predictions, axis=1).numpy()
print("Predicted labels:", predicted_labels)
print("True labels:", y_test[:5])

# Visualize predictions
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(x_test[i].reshape(28, 28, 1), cmap='OrRd')
    plt.title(f"Predicted: {predicted_labels[i]}\nTrue: {y_test[i]}")
    plt.axis('off')
plt.show()

# plot training and validation accuracy
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title("Accuracy over Epochs")
plt.show()

# plot training and validation loss
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title("Loss over Epochs")
plt.show()
```
